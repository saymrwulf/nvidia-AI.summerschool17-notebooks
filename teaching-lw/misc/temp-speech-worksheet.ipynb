```python
def read_lab(fname):    result = []    with open(fname) as stream:        for line in stream.readlines():            line = line[:-1]            start,stop,label = line.split(" ",2)            result.append((float(start),float(stop),label))    return resultread_lab("msak0_306.lab")
```

```python
import osclass Wave:    passdef read_data(fname):    os.system("sox %s _temp.raw"%fname)    data = numpy.fromfile("_temp.raw",'int16')    wave = Wave()     base,_ = os.path.splitext(fname)    wave.labels = read_lab(base+".lab")    wave.data = data    wave.rate = 16000.0    return wavewave = read_data("msak0_306.wav")
```

```python
def phonemes(wave):    for start,stop,label in wave.labels:        start = int(wave.rate*start)        stop = int(wave.rate*stop)        #print start,stop,len(wave.data)        clip = wave.data[start:stop]        yield clip,labelclips = list(phonemes(wave))
```
