Pencil Sketch=============Here is a simple solution to the pencil sketch problem. There are many other ways of getting "pencil sketch"-like effects.For this effect, we just use linear filters, per-pixelarithmetic, and noise.

```python
from pylab import *gray()def rescale(image):    """Rescale the pixels of an image into the range 0...1"""    return (image-amin(image))/(amax(image)-amin(image))
```

```python
image = imread("Professor_X_03.jpg")bw = mean(image,axis=2)from scipy.ndimage import filtersbw = rescale(bw)figure(figsize=(8,8)); imshow(bw,origin='lower')
```

It turns out that a bit of median filtering may make this work a little nicer,but it's not needed.  Experiment if you like.

```python
# bw = filters.median_filter(bw,3)# figure(figsize=(8,8)); imshow(bw,origin='lower')
```

We want some kind of texture that suggests pencil-based shading.If you look at some of these, they tend to be streaked diagonally.A little experimentation shows that we can simulate this byadding noise along a diagonal.Alternatively, you can simply take a real uniformly shadedpencil sketch and just scan that as a texture (there aresome of those images around the web).

```python
noise = rescale(clip(randn(*bw.shape),-0.0,2.0))streaked = zeros(noise.shape)for i in range(-7,8): streaked += roll(roll(noise,i,0),i,1)streaked /= amax(streaked)imshow(0.7+0.3*streaked,origin='lower',vmin=0,vmax=1)
```

The above texture already is pretty good, but perhaps a little too uniform.  We can add a somewhat coarser texture to it by smoothing it and addingit to itself.

```python
from scipy.ndimage import filterssmoothed = filters.gaussian_filter(streaked,1.0)smoothed /= amax(smoothed)combo = 0.7*smoothed+0.3*streakedimshow(0.7+0.3*combo,origin='lower',vmin=0,vmax=1)
```

Simply adding this to the raw image doesn't work well.  What we wantis to "modulate" the image with the pencil, we want to use the pencilstrokes to render the image.  Where there are no pencil strokes, thebackground should appear.This suggests multiplication, not addition, and that works pretty well.

```python
result = rescale((0.4*combo+0.8)*bw)figure(figsize=(8,8)); imshow(result,origin='lower',vmin=0,vmax=1)
```

Actually, that image is still a little too contrasty; with a real pencil, you don't achieve that kind of contrast.So let's tinker with the contrast some.

```python
result = 0.2+0.8*rescale((0.4*combo+0.8)*(0.5+0.5*bw))figure(figsize=(8,8)); imshow(result,origin='lower',vmin=0,vmax=1)
```

Not perfect, but reasonably good.

Edge Enhancements=================In pencil sketches, artists often emphasize edges before filling in regions.  We can simulate that as well using the gradient filters wediscussed.A little experimentation shows that clipping the gradient values is a good idea, since otherwise only the strongest gradients show up.First, we need to compute the gradient magnitude.For display purposes, let's show this in reverse colors.The background noise gives us some nice extra texture.

```python
grad = rescale(filters.gaussian_gradient_magnitude(bw,1.0)**0.7)grad = rescale(clip(grad,0,amax(grad)*0.4))figure(figsize=(8,8)); imshow(1-grad,origin='lower',vmin=0,vmax=1)
```

Now we mix this with the original grayscale image.

```python
mix = (1.0-grad**0.3)*0.5+bw*0.5figure(figsize=(8,8)); imshow(mix,origin='lower')
```

Finally, let's apply penciling to this image again. 

```python
result = 0.2+0.8*rescale((0.4*combo+0.8)*(0.4+0.6*mix))figure(figsize=(8,8)); imshow(result,origin='lower',vmin=0,vmax=1)
```

Strictly speaking, we should apply the pencil strokes only to the shaded areas, not to the edges, so you can try to change thealgorithm accordingly.If you want something better than this, you probably have to use some kind of anisotropic filtering (from later lectures).
