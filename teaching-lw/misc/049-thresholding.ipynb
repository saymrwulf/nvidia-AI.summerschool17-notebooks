Simple Thresholding
===================

```python
subplots(1,1,figsize=(8,8))
image = mean(imread("Initialen.jpg"),axis=2)
gray(); imshow(image,origin='lower')
```

```python
imshow(image[200:400,200:400],origin='lower')
```

```python
_=hist(image.ravel(),bins=100)
```

```python
subplots(1,1,figsize=(8,8))
imshow(image>130,origin='lower')
```

```python
imshow(image[200:400,200:400]>130,origin='lower')
```

Adaptive Thresholding with Linear Filtering
=====================

```python
subplots(1,1,figsize=(8,8))
image = mean(imread("bookpage.jpg"),axis=2).T/255.0
gray(); imshow(image)
```

```python
_=hist(image.ravel(),bins=100)
```

```python
subplots(1,1,figsize=(8,8))
imshow(image>0.5)
```

```python
subplots(1,1,figsize=(8,8))
from scipy.ndimage import filters
avg = filters.gaussian_filter(image,100.0)
imshow(avg)
```

```python
subplots(1,1,figsize=(16,8))
nobg = image-avg
subplot(121); imshow(image); subplot(122); imshow(nobg)
```

```python
_=hist(nobg.ravel(),bins=100)
```

```python
subplots(1,1,figsize=(8,8))
thresholded = (nobg>-0.1)
imshow(thresholded,origin='lower')
```

```python
subplot(121); imshow(thresholded[500:600,500:600]); subplot(122); imshow(thresholded[1900:2000,1200:1300])
```

Variance Estimation with Linear Filters
===================

```python
avg = filters.gaussian_filter(image,100.0)
avg2 = filters.gaussian_filter(image**2,100.0)
var = avg2 - avg**2
subplots(1,1,figsize=(8,8))
imshow(var)
```

Threshold Estimation with Median Filters
========================================

```python
mf = filters.median_filter(image,50)
subplots(1,1,figsize=(8,8))
imshow(mf)
```

```python
subplots(1,1,figsize=(8,8))
imshow(image-mf)
```

```python
_=hist((image-mf).ravel(),bins=100)
```

```python
subplots(1,1,figsize=(8,8))
thresholded = (image-mf>-0.1)
imshow(thresholded)
```

```python
subplot(121); imshow(thresholded[500:600,500:600]); subplot(122); imshow(thresholded[1900:2000,1200:1300])
```

Threshold Estimation with Grayscale Morphology
==============================================

```python
from scipy.ndimage import morphology
go = morphology.grey_closing(image,size=(50,50))
subplots(1,1,figsize=(8,8))
imshow(go)
```

```python
ng = image-go+1.0
subplots(1,1,figsize=(8,8))
imshow(ng)
```

```python
thresholded = (ng>0.8)
subplots(1,1,figsize=(8,8))
imshow(thresholded)
```

```python
subplot(121); imshow(thresholded[500:600,500:600]); subplot(122); imshow(thresholded[1900:2000,1200:1300])
```
