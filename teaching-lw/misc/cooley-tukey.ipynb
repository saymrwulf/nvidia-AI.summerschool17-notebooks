```python
from pylab import *from scipy.ndimage import filters
```

```python
s = filters.gaussian_filter(randn(1024),2.0)plot(s)
```

```python
X = fft(s)E = fft(s[::2])O = fft(s[1::2])
```

```python
plot(abs((X[:512]-E)/O))plot(angle((X[:512]-E)/O))
```

```python
plot(abs((X[512:]-E)/O))plot(angle((X[512:]-E)/O))
```
