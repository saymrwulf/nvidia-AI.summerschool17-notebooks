===============

```python
with tables.openFile("mnist.h5","r") as mnist:
    print mnist
    images = array(mnist.root.deskewed,'f')
    labels = array(mnist.root.labels,'i')
    test_images = array(mnist.root.test_deskewed,'f')
    test_labels = array(mnist.root.test_labels,'i')
```

```python
figsize(6,6)
mu = mean(images,axis=0)
gray()
imshow(mu)
```

Let's start by centering the images.

```python
centered = images - mu[newaxis,:,:]
```

Next, we start off with some random vectors as our initial guesses for the eigenvectors.

```python
d = 28*28
evs = array([randn(d) for i in range(6)])
evs = array([e/norm(e) for e in evs])
```

The key idea behind the algorithm is the following.

We the first eigenvector is reached in the limit as $M^k x / ||M^k x||$
for some random starting vector $x$.
Actually, for the computation, it doesn't really matter when we renormalize $x$.
We can perform some matrix multiplications and then normalize if we like.

$M$ is the covariance matrix $\frac{1}{N}\sum x_i^Tx$.
So, we get $Mx = \frac{1}{N}(\sum x_ix_i^T)x=\frac{1}{N}\sum x_i(x_i\cdot x)$.

Now, if the training vectors $x_i$ are actually just i.i.d. samples from some
distribution, then it really doesn't matter that much how large we choose $N$.  All
we are really doing is computing an average over a chunk of them and then multiplying
with $x$--or exchaning the multiplications.  And it also doesn't really matter when
we renormalize $x$.

To arrive at a fully incremental algorithm, we replace the average with
a moving average.  This amounts to giving the samples different amounts of weight.

```python
eta = 0.001
def update(x):
    x = x/norm(x)
    for i,e in enumerate(evs):
        e += eta*x*dot(x,e)
        e /= norm(e)
        x -= dot(e,x)*e
```

Now we just perform a lot of updates.
With the moving average and all this isn't all that
efficient, but it requires very little storage
(just the eigenvectors themselves).

```python
for iter in range(1000000):
    if iter%100000==0: print iter
    update(centered[iter%len(centered)].ravel())
```

Let's look at the eigenvectors now.  When you look at the eigenvectors of the regular PCA,
you see that these are very similar (some of them have their signs switched).

```python
figsize(10,10)
for i in range(6):
    subplot(1,6,i+1)
    imshow(evs[i].reshape(28,28),cmap=cm.jet)
```
